```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('../data/raw/zomato.csv')
```

```python
df.shape
```

```python
df = df[df['url'].str.startswith('http', na=False)]
```

```python
df.shape
```

```python
df = df[[
    'name',
    'online_order',
    'book_table',
    'rate',
    'votes',
    'location',
    'rest_type',
    'cuisines',
    'approx_cost(for two people)',
    'listed_in(type)',
    'listed_in(city)'
]]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rest_type'] = df['rest_type'].fillna('unknown')
```

```python
df = df.dropna(subset=[
    'name',
    'location',
    'cuisines',
    'rate',
    'approx_cost(for two people)'
])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df[df['rate'].notnull()]
df = df[~df['rate'].isin(['NEW', '-'])]

df['rate'] = df['rate'].str.replace('/5', '', regex=False)
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

df = df[df['rate'].notnull()]
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=['name', 'location'])
```

```python
df.shape
```

```python
df['location'] = df['location'].str.lower().str.strip()
df['cuisines'] = df['cuisines'].str.lower().str.strip()
df['rest_type'] = df['rest_type'].str.lower().str.strip()
```

```python
df.head()
```

```python
df = df.rename(columns={
    'rate': 'rating',
    'approx_cost(for two people)': 'cost',
    'listed_in(type)': 'service_type',
    'listed_in(city)': 'city'
})
```

```python
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df['restaurant_id'] = ['R' + str(i) for i in range(1, len(df) + 1)]
```

```python
df.head()
```

```python
df = df[[
    'restaurant_id',
    'name',
    'location',
    'rest_type',
    'cuisines',
    'rating',
    'votes',
    'cost',
    'online_order',
    'book_table',
    'service_type',
    'city'
]]
```

```python
df.head()
```

```python
df.describe()
```

```python
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0})
df['book_table'] = df['book_table'].map({'Yes': 1, 'No': 0})
```

```python
df['cost'].dtype
```

```python
df['cost'] = df['cost'].astype(str)

# remove commas if any
df['cost'] = df['cost'].str.replace(',', '')

# convert to numeric
df['cost'] = pd.to_numeric(df['cost'], errors='coerce')
```
